```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("daily_temperature.csv")  # Ensure columns: Day, Temperature

# Plot line chart
plt.figure(figsize=(10,5))
plt.plot(df["Day"], df["Temperature"], marker='o', linestyle='-', color='b', label="Temperature")

# Add grid, labels, and title
plt.grid(True, linestyle="--", alpha=0.6)
plt.xlabel("Day")
plt.ylabel("Temperature (C)")
plt.title("Daily Temperature Variation Over a Month")
plt.legend()

# Show chart
plt.show()

# Find hottest and coldest days
hottest_day = df.loc[df["Temperature"].idxmax()]
coldest_day = df.loc[df["Temperature"].idxmin()]

print(f" Hottest Day: Day {hottest_day['Day']} with {hottest_day['Temperature']}C")
print(f" Coldest Day: Day {coldest_day['Day']} with {coldest_day['Temperature']}C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_data.csv")  # Ensure columns: Quarter, Sales

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df["Quarter"], df["Sales"], marker='o', linestyle='-', color='g', label="Sales")

# Highlight peak sales
peak = df.loc[df["Sales"].idxmax()]
plt.scatter(peak["Quarter"], peak["Sales"], color="red", s=100, zorder=5, label="Peak Sales")

# Add grid, labels, and title
plt.grid(True, linestyle="--", alpha=0.6)
plt.xlabel("Quarter")
plt.ylabel("Sales (in  or $)")
plt.title("Quarterly Sales Performance")
plt.legend()

# Show chart
plt.show()

print(f" Peak Sales in {peak['Quarter']} with {peak['Sales']} units/revenue.")
```
